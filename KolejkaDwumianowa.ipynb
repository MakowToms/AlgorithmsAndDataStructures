# Kolejki priorytetowe - kopce złączalne
## Kolejka Dwumianowa
### Operacje:
- union (łączenie 2 kolejek)
- insert (dodawanie nowego elementu)
- max (pobieranie maksymalnego elementu)
- deleteMax (pobieranie i usuwanie maksymalnego elementu)
- increaseKey (zwiększanie konkretnego klucza)

```python
import math
def mergeTrees(t1,t2):
    '''
    Łączenie drzew o takiej samej wysokości (na wejściu otrzymujemy 2 nody - elementy kolejki dwumianowej)
    '''
    if t1.key < t2.key:
        t1,t2 = t2,t1
    if t1.child == None:
        t1.child = [t2]
    else:
        t1.child.append(t2)
    t1.rank = t1.rank + 1
    return t1

def mergeLists(l1,l2):
    '''
    Łączenie 2 list z nodami
    '''
    merged = []
    while len(l1) > 0 or len(l2) > 0:
        if len(l1) == 0:
            merged.append(l2.pop(0))
        elif len(l2) == 0:
            merged.append(l1.pop(0))
        elif l1[0].rank > l2[0].rank:
            merged.append(l2.pop(0))
        else:
            merged.append(l1.pop(0))
    return merged

class BinomialHeap:
    class Node:
        def __init__(self,key):
            self.key = key
            self.rank = 0
            self.child = None
    
    def __init__(self,heap=None):
        self.head = heap
    
    def union(self,other):
        if self.head == None:
            self.head = other.head
            return
        if other.head == None:
            return
        merged = mergeLists(self.head,other.head)
        i = 0
        while i<len(merged):
            if i+1<len(merged) and merged[i].rank == merged[i+1].rank:
                if i+2<len(merged) and merged[i+1].rank == merged[i+2].rank:
                    new = mergeTrees(merged[i+1],merged[i+2])
                    merged.pop(i+1)
                    merged.pop(i+1)
                    merged.insert(i+1,new)
                    i += 1
                    continue
                new = mergeTrees(merged[i],merged[i+1])
                merged.pop(i)
                merged.pop(i)
                merged.insert(i,new)
                continue
            i += 1
        self.head = merged
    
    def insert(self,key):
        new = BinomialHeap()
        new.head = [BinomialHeap.Node(key)]
        self.union(new)
    
    def deleteMax(self):
        if len(self.head) == 0:
            return None
        M = -math.inf
        i_max = 0
        for i in range(len(self.head)):
            if self.head[i].key > M:
                M = self.head[i].key
                i_max = i
        to_throw = self.head.pop(i_max)
        new = BinomialHeap()
        new.head = to_throw.child
        self.union(new)
        return to_throw.key
```

```python
import random
binomialHeap = BinomialHeap()
for i in range(10):
    binomialHeap.insert(random.randint(1,20))
for i in range(11):
    print(i,binomialHeap.deleteMax())
```
